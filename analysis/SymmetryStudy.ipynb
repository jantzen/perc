```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
files = glob.glob('../data/topsoil*')
files.extend(glob.glob('../data/sand_[0-9]*'))
```

Make data frames:

```python
frames = []
for f in files:
    frames.append(pd.DataFrame((np.loadtxt(f)).T,columns=["time","volume","pressure"]))
```

```python
print(len(frames))
```

```python
frames[0].head()
```

```python
for ii, f in enumerate(files):
    print(ii, f)
```

Smooth the data sets:

```python
smooth = []
span=100
for fr in frames:
    smooth.append(fr['volume'].ewm(span=span).mean())
```

```python
smooth[0].head()
```

Convert to volume.

```python
vol = []
for sm in smooth:
    tmp = sm.to_numpy()
    tmp = tmp[span:]
    tmp = (tmp[0] - tmp) / 10. * np.pi * (2.125 * 2.54)**2
    vol.append(tmp)
```

```python
vol[0].head()
```

```python
for v in vol:
    plt.plot(v[:50])
```

```python
vol_clipped = []
for v in vol:
    #sm = series.to_numpy()
    tmp = v[:np.min(np.where(v>1000.))]
    print(len(tmp))
    vol_clipped.append(tmp)
```

```python
plt.figure()
for vc in vol_clipped:
    plt.plot(vc)
```

Split each data set into two (untransformed and transformed):

```python
syms = []
for vc in vol_clipped:
    cut = np.min(np.where(vc>460.))
    untrans = vc[:cut]
    trans = vc[cut:(cut+len(untrans))]
    syms.append([untrans,trans])
print(len(syms))
```

```python
plt.figure()
for s in syms:
    plt.plot(s[0],s[1])
```

```python
fits=[]
for s in syms:
    f = np.polyfit(s[0],s[1],1)
    fits.append(np.array(f).reshape(1,-1))
fits = np.concatenate(fits,axis=0)
print(np.mean(fits, axis=0))
```
